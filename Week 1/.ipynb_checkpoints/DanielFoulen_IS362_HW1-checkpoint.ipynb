```python
# =========================
# HW 1 – Python Basics
# =========================

print("QUESTION 1: Combine variables\n")

one = 1
two = 2
three = 3
four = 4
can = "can"
do = "do"
python = "python"
more = "more"

print(one, two, three, four, can, do, python, more)
```

```python
# =========================
print("\nQUESTION 2: Multiply numbers\n")

number1 = 15
number2 = 30

result = number1 * number2
print("The result is", result)
```

```python
# =========================
print("\nQUESTION 3: First and last number comparison\n")

def first_last_same(numbers):
    return numbers[0] == numbers[-1]

numbers_x = [10, 20, 30, 40, 10]
numbers_y = [75, 65, 35, 75, 30]

print("Given list:", numbers_x)
print("result is", first_last_same(numbers_x))

print("Given list:", numbers_y)
print("result is", first_last_same(numbers_y))
```

```python
# =========================
print("\nQUESTION 4: Numbers divisible by 5\n")

numbers = [10, 20, 33, 46, 55]

print("Given list is", numbers)
print("Divisible by 5")

for num in numbers:
    if num % 5 == 0:
        print(num)
```

```python
# =========================
print("\nQUESTION 5: Count substring occurrences\n")

str_x = "Anjali is good developer. Anjali is a writer"
count = str_x.count("Anjali")

print("Anjali appeared", count, "times")
```

```python
# =========================
print("\nQUESTION 6: Print number pattern\n")

for i in range(1, 6):
    print(str(i) * i)
```

```python
# =========================
print("\nQUESTION 7: Multiplication table (1 to 10)\n")

for i in range(1, 11):
    for j in range(1, 11):
        print(i * j, end="\t")
    print()
```

```python
# =========================
print("\nQUESTION 8: NumPy – Kelvin to Celsius\n")

import numpy as np

Temp_K = [390.65, 107.52, 225.25, 275.98, 318.18, 88.45]
temp_array = np.array(Temp_K)

temp_celsius = temp_array - 273.15
print("Temperatures in Celsius:")
print(temp_celsius)
```

```python
# =========================
print("\nQUESTION 9: Pandas DataFrame\n")

import pandas as pd

data = {
    "Student": ["Tommy", "Fred", "Gail", "Daniel", "Randy", "Ariel"],
    "Hair Color": ["Brown", "Grey", "Blonde", "Brown", "Black", "Brown"],
    "Eye Color": ["Hazel", "Green", "Blue", "Brown", "Hazel", "Blue"],
    "Weight (kg)": [60, 80, 76, 70, 88, 85]
}

df = pd.DataFrame(data)

print("Full DataFrame:")
print(df)

print("\nHair Color column (Series):")
print(df["Hair Color"])

print("\nEye Color column (DataFrame):")
print(df[["Eye Color"]])

print("\nStudent and Weight columns:")
print(df[["Student", "Weight (kg)"]])

print("\nSecond through fourth observations:")
print(df.iloc[1:4])
```

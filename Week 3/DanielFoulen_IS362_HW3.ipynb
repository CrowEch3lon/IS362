Suppose I rode a bicycle from New York City to Key West, Florida. Each day I recorded my bicycle odometer reading, which shows the **cumulative** total miles ridden so far.

This notebook stores those cumulative readings in a pandas Series and computes the **daily miles ridden** from them.

## Cumulative vs Daily Miles

The odometer records the total miles ridden at the end of each day. For example, if the first two readings are `55` and `120`, then:

- Day 1 miles = `55`
- Day 2 miles = `120 - 55 = 65`

So to recover daily mileage, we compute the difference between consecutive cumulative values.


```python
import pandas as pd
```

## Cumulative Odometer Readings (End of Each Day)

Below is an example set of cumulative miles recorded over 7 days.


```python
cumulative_miles = pd.Series(
    [55, 120, 190, 250, 310, 380, 450],
    name="Cumulative Miles"
)

cumulative_miles
```

## Daily Miles Ridden

To compute miles ridden each day, we take the difference between each day’s cumulative value and the previous day’s cumulative value.

In pandas, `.diff()` computes this difference. The first day has no previous value, so we replace the missing first result with the first day’s cumulative miles.


```python
daily_miles = cumulative_miles.diff()
daily_miles.iloc[0] = cumulative_miles.iloc[0]
daily_miles.name = "Miles Ridden That Day"

daily_miles
```

## Presenting the Results

The table below shows both the cumulative miles (odometer reading) and the miles ridden each day side by side.


```python
summary = pd.DataFrame({
    "Cumulative Miles": cumulative_miles,
    "Miles Ridden That Day": daily_miles
})

summary.index = summary.index + 1
summary.index.name = "Day"

summary
```

## Conclusion

By storing cumulative odometer readings in a pandas Series and taking differences between consecutive days, we can recover the total miles ridden each day in a clear and reproducible way.
